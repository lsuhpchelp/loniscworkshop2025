# This Jupyter notebook will help you understand the gradient descent process by finding the minimum value of a given object function $f(x)=(x-1)^2$ within the range of $[0,2]$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rc
#rc('animation', html='jshtml')
```

## This is the object function we want to minimize $f(x)=(x-1)^2$, the argument is a vector, so think about what numpy function we are using?

```python
# object function
def f(x):
    return (x-1)**2;
```

## Try to fill in the derivative of $f(x)$:

## Solution
Check the below cell for solution:

```python
# calculate the derivative of f(x)
def df(x):
    # what is the derivative of f(x) ?
    return 2*(x-1);
```

## Here we define the gradient descent function:

```python
def gradient_descent(start_at=np.random.uniform(-1,1),
                     iterations=20,
                     learning_rate=0.1):
    # record all intermediate values
    x_vals, y_vals = [], []
    print("start_at=", start_at)
    x = start_at
    x_vals.append(x)
    y_vals.append(f(x))
    for _ in range(iterations):
        gradient = df(x) # derivative of f(x)
        x += learning_rate * -gradient
        y = f(x)
        x_vals.append(x)
        y_vals.append(y)
        #plt.scatter(x, y, color='grey', edgecolor='black', zorder=3)
    return x_vals, y_vals
```

## Call the gradient descent function from here:

```python
xv, yv = gradient_descent(start_at = np.random.uniform(2,2), iterations = 30, learning_rate=0.1)
#print(xv)
#print(yv)
```
